### 1.3

```python
three_consecutive_vowels = re.compile(r'\b\w*[b-df-hj-np-tv-z]+[aeiou]{3}[b-df-hj-np-tv-z]+\w*\b|\b[aeiou]{3}[b-df-hj-np-tv-z]+\w*\b',re.IGNORECASE)
three_consecutive_vowels_list = []
for string in string_list:
    z = re.search(three_consecutive_vowels,string)
    if z:
        three_consecutive_vowels_list.append(z.group())
len(three_consecutive_vowels_list)
```

```python
re.findall(three_consecutive_vowels, 'gsuGeeegzxgj\nahgasdee\ngeefeeeefe\nEeefg')
```

## 2 Exploring Internet Traffic with Regexs
### 2.1

```python
response = ur.urlopen("http://umich.edu/~klevin/teaching/Fall2019/STATS507/SkypeIRC.txt")
read = response.read()
string = read.decode("utf-8")
str_packets = re.split(r'\n+',string)
for st in str_packets:
    if len(st) == 0:
        str_packets.remove(st)
n_packets = len(str_packets)
for st in str_packets:
    print(st + '\n')
```

```python
n_packets
```

There are 2263 many packets.

### 2.2

```python
ip_address = re.compile(r'(([0-1]?[0-9][0-9]?|2[0-4][0-9]|25[0-5])\.){3}([0-1]?[0-9]?[0-9]?|2[0-4][0-9]|25[0-5]){1}')
ip_addresses = []
for st in str_packets:
    l = re.split(r'\s',st)
    for r in l:
        z = re.search(ip_address,r)
        if z:
            ip_addresses.append(z.group())
len(ip_addresses)
print(ip_addresses)
```

### 2.3

```python
def get_packets_by_regex(input_string):
    if type(input_string) is not str:
        raise ValueError('Not a string')
    op_lines = []
    for line in str_packets:
        z = re.search(input_string,line)
        if z:
            op_lines.append(line)
    return op_lines
```

```python
get_packets_by_regex(r'comcast')
```

```python
time_regex = re.compile(r'\s+(\d+)\s+(\d+\.?\d*)\s')
```

```python
str_lines = re.split(r'\n',string)
time = []
for line in str_lines:
    z = re.search(time_regex,line)
    if z:
        time.append(z.groups()[1])
print(time)
```

### 2.4

```python
import numpy as np
import matplotlib.pyplot as plt
time_int = np.array(time,dtype = 'float32')
time_int = np.floor(time_int)
plt.hist(time_int,bins = 50)
plt.savefig('timestamps.pdf')
```

## 3 Retrieving Data from the Web
### 3.1

```python
import urllib.request
from bs4 import BeautifulSoup
response = urllib.request.urlopen('https://en.wikipedia.org/wiki/Ann_Arbor,_Michigan')
parsed = BeautifulSoup(response.read(), 'html.parser')
reg = re.compile(r'Climate data for[ \w]+')
table_list = []
for table in parsed.find_all('table'):
    z = re.search(reg,table.get_text())
    if z:
        table_list.append(table)
print(table_list[0].prettify())
```

For the city of Ann Arbor, Michigan, the climate table has one 'tbody' node, under which we have 12 nodes for table rows ('tr'); except 1st, 2nd and last rows, all the other rows have one table header ('th') tag and 13 standard cell tags ('td').

```python
import urllib.request
from bs4 import BeautifulSoup
import pandas as pd

## Function for retiving climate data from Wikipedia of a city
def retrieve_climate_table(web_address):
    response = urllib.request.urlopen(web_address) #
    if response.getcode() != 200:
        raise HttpError("Http error code:" + str(response.getcode()) + '\n')
    parsed = BeautifulSoup(response.read(), 'html.parser')
    reg = re.compile(r'Climate data for[ \w+]+')
    table_list = []
    for table in parsed.find_all('table'):
        z = re.search(reg,str(table))
        if z:
            table_list.append(table)
    if len(table_list):
        return table_list[0]
    else:
        raise ValueError('No climate table found')
retrieve_climate_table('https://www.wtdgsjhds.com')
```

```python
def list_climate_table_row_name(web_address):
    table = retrieve_climate_table(web_address)
    row_list = []
    for h in table.find_all('th'):
        if h.get('scope') == 'row':
            string = h.get_text()[:-1]
            row_list.append(string)
    return row_list
list_climate_table_row_name('https://en.wikipedia.org/wiki/Harbin')
```
